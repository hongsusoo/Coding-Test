```python
def tilt(p_r,p_b,xx,yy,ma,move=0):
    #R이동
    d_r=p_r+m*yy+xx
    d_b=p_b+m*yy+xx
    if ma[d_r]=='.':
        ma[p_r] = '.'
        ma[d_r] = 'R'
        return tilt(d_r,p_b,xx,yy,ma,move)
    elif ma[d_r]=='O' and move==0: #구슬빠짐
        move=1
        ma[p_r] = '.'
        return tilt(p_r,p_b,xx,yy,ma,move)
    elif ma[d_b]=='.':
        ma[p_b] = '.'
        ma[d_b] = 'B'
        return tilt(p_r,d_b,xx,yy,ma,move)
    elif move==1 and ma[d_b]!='O': return (p_r,p_b,ma,True)
    else : return (p_r,p_b,ma,False)

from collections import deque

# n=3
# m=10
# ma="###########.O....BR###########"
# n=5
# m=5
# ma="######..B##.#.##RO.######"
# n=7
# m=7
# ma="########...RB##.######.....######.##O....########"

n=3
m=7
ma="########B#O#R........"
# n=10
# m=10
# ma="###########R#...##B##...#.##.######.##.##......#.##.######.##.#....#.##.#.#.#..##...#.O#.###########"
# n=10
# m=10
# ma="###########R#...##B##...#.##.######.##.##......#.##.######.##.#....#.##.#.##...##O..#....###########"

xx=(0,0,1,-1)
yy=(1,-1,0,0)
z=ma.find("O")
p_r=ma.find("R") #R 위치
p_b=ma.find("B") #B 위치
ma=list(ma)
q=deque()
c=0
s=set() #DP 동적 프로그래밍(중복제거)
t=False
q.append((p_r,p_b,c,ma,s))

while q:
    p_r,p_b,c,ma,s = q.popleft()
    s.add((p_r,p_b))
    if c>10:print(-1); break
    for i in range(4):
        dpr,dpb,ma,t = tilt(p_r,p_b,xx[i],yy[i],ma)
        print(s,(dpr,dpb),not (dpr,dpb) in s)
        if t==True: break
        if not (dpr,dpb) in s:
            q.append((dpr,dpb,c+1,ma,s))
    if t==True: print(c+1); break
else: print(-1)
```

```python
def move(r,b,s,c,i):
    rx,ry=r%m,r//m
    bx,by=b%m,b//m
    if i==0: #up
        ry-=1
        by-=1
    elif i==1:#Down
        ry+=1
        by+=1
    elif i==2:#right
        rx+=1
        bx+=1
    elif i==3:#left
        rx-=1
        bx-=1
#     print(rx,ry)    
    if 0 <= rx < m and 0 <= ry < n and 0 <= bx < m and 0 <= by < n:
        rr=m*ry+rx
        bb=m*by+bx  
        print(ma[rr])
        if ma[rr]=="O":
            return True
        elif ma[bb]=="O":
            return c
        else :
            if ma[rr]=="." and ma[bb]==".": #둘다 움직일 수 있을때 
                move(rr,bb,s,c,i)
            if ma[rr]=="." and ma[bb]!=".": #R만 움직일 수 있는 상태
                move(rr,b,s,c,i)
            if ma[rr]!="." and ma[bb]==".": #B만 움직일 수 있는 상태
                move(r,bb,s,c,i)
            else : 
                if r not in s:
                    s=s.add(r)
                    return (r,b,s,c)


from collections import deque

n=3
m=10
ma="###########.O....RB###########"

p_r=ma.find("R") #R 위치
p_b=ma.find("B") #B 위치
s_r=set()
s_r.add(p_r) 
c=0
t=11
q=deque()
q.append((p_r,p_b,s_r,c))

while q:
    r,b,s,c=q.popleft()
#     print(q)
#     print(t,c)
    if c>t: print(-1); break
    for i in range(4): # 4방향 움직이면서 Q추가
        mm=move(r,b,s,c+1,i)
#         print(mm,c)
        if mm==True: print(c+1); break
        elif mm==c+1: t=c
        elif mm: q.append(mm)      
    if c>=10:
        print(-1)
        break
```
