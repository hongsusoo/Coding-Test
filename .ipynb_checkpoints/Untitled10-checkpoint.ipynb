```python
for _ in range(int(input())):
    sm, sc, si = map(int,input().split())
    pgr_code = list(input())
    input_str = list(map(ord,input()))

    mat_bracket = {}
    temp=[]
    for i in range(sc):
        if pgr_code[i]=='[':
            temp.append(i)
        elif pgr_code[i]==']':
            mat_bracket[i]=temp[-1]
            mat_bracket[temp.pop()]=i
     
    cnt = 0
    pointer = 0
    pos_code = 0
    pos_instr = 0
    sys_memo = [0]*sm 
    
    while cnt<=100000000:
        cnt+=1
        c = pgr_code[pos_code]
        if c=='+':
            sys_memo[pointer] = sys_memo[pointer]+1 if sys_memo[pointer] !=255 else 0
        elif c=='-':
            sys_memo[pointer] = sys_memo[pointer]-1 if sys_memo[pointer] !=0 else 255
        elif c=='>':
            pointer = pointer+1 if pointer<sm else: 0
        elif c=='<':
            pointer = pointer-1 if pointer>0 else: sm-1
        elif c=='[':
            if sys_memo[pointer] == 0:
                pos_code = mat_bracket[pos_code]+1
        elif c==']':
            if sys_memo[pointer] != 0:
                pos_code = mat_bracket[pos_code]+1
        elif c==',':
            if pos_instr<si:
                sys_memo[pointer] = input_str[pos_instr]
                pos_instr+=1
            else :
                sys_memo[pointer] = 255
        pos_code+=1
        
        
```

```python
for _ in range(int(input())):
    sm, sc, si = map(int, input().split())
    code = list(input())
    input_str = list(map(ord, input())) + [255]
	
    # 대괄호 점프할 곳 찾기
    jump = {}
    temp = []
    for i in range(sc):
        if code[i] == '[':
            temp.append(i)
        elif code[i] == ']':
            start = temp.pop()
            jump[start] = i
            jump[i] = start

    memory = [0] * sm
    pm = 0
    pc = 0
    pi = 0
    cnt = 0
    stack = []  # 루프 깊이 저장
    while cnt < 1000:
        cnt += 1
        print("code :"+code[pc])
        if pc >= sc:  # 프로그램 끝
            break  # Terminates
        if code[pc] == '-':
            memory[pm] = (memory[pm] - 1) % 256
            pc += 1
        elif code[pc] == '+':
            print(memory[pm])
            memory[pm] = (memory[pm] + 1) % 256
            pc += 1
        elif code[pc] == '<':
            pm = (pm - 1) % sm
            pc += 1
        elif code[pc] == '>':
            pm = (pm + 1) % sm
            pc += 1
        elif code[pc] == '[':
            if memory[pm] == 0:
                pc = jump[pc] + 1
            else:
                stack.append(pc)
                pc += 1
        elif code[pc] == ']':
            if memory[pm] != 0:
                pc = jump[pc] + 1
            else:
                stack.pop()
                pc += 1
        elif code[pc] == '.':
            pc += 1
        elif code[pc] == ',':
            memory[pm] = input_str[pi]
            print(memory[pm],',',input_str[pi])
            if pi < si:
                pi += 1
            pc += 1
    else:  # 5천만번동안 종료x
        print('Loops {} {}'.format(stack[0], jump[stack[0]]))  # loop
        continue
    print('Terminates')  # Terminates
```
