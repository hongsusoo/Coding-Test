```python
def cng_arr(n,m,t_arr,odr):
    r=[]
    if odr == 1: # 상하 반전
        for i in range(n):
            r=[t_arr[i]]+r
        return r
    if odr == 2: #좌우 반전
        for i in range(n):
            r=r+[t_arr[i][::-1]]
        return r
    if odr ==3: #오른쪽 90도 회전
        r = [[t_arr[i][j] for i in range(n-1,-1,-1)] for j in range(m)]
        return r
    
    if odr == 4: #왼쪽 90도 회전
        r = [[t_arr[i][j] for i in range(n)] for j in range(m-1,-1,-1)]
        return r

#입력 받기
arr = []
N,M,T = map(int,input().split())
for _ in range(N):
    arr += (list(map(int,input().split())))
order = list(map(int,input().split()))


A = [arr[i*M:i*M+int(M/2)] for i in range(int(N/2))]
B = [arr[i*M+int(M/2):i*M+M] for i in range(int(N/2))]
C = [arr[i*M+int(N/2)*M:i*M+int(N/2)*M+int(M/2)] for i in range(int(N/2))]
D = [arr[i*M+int(N/2)*M+int(M/2):i*M+M+int(N/2)*M] for i in range(int(N/2))]

seq=[A,B,C,D]
#기준좌표
coor = {0:(0,1),1:(1,0),2:(0,-1),3:(-1,0)}
x=0
y=1
for i in order:
    a,b,c,d = seq
    if i == 1: #상하 반전
        seq = [c,d,a,b]
        x=(~x&1)*(x^(0^2))+(x&1)*x #짝수면 반대수 출력
        y=(~y&1)*(y^(0^2))+(y&1)*y #홀수면 그대로 출력
    elif i == 2: #좌우 반전
        seq = [b,a,d,c]
        x=(x&1)*(x^(1^3))+(~x&1)*x #짝수면 그대로 출력
        y=(y&1)*(y^(1^3))+(~y&1)*y #홀수면 반대수 출력
    elif i in (3,5): #오른쪽 90도 회전
        seq = [c,a,d,b]
        if i == 3:
            x=(x+1)%4
            y=(y+1)%4
    elif i in (4,6): #왼쪽 90도 회전
        seq = [b,d,a,c]
        if i==4:
            if x==0 : x=3
            else: x-=1
            if y==0 : y=3
            else: y-=1

fin_order = [3]*x

if y-x in (-1, 3): 
    if y in (0,2): fin_order+=[1]
    else : fin_order+=[2]
        
n=int(N/2)
m=int(M/2)

for o in fin_order:
    seq[0]=cng_arr(n,m,seq[0],o)
    seq[1]=cng_arr(n,m,seq[1],o)
    seq[2]=cng_arr(n,m,seq[2],o)
    seq[3]=cng_arr(n,m,seq[3],o)
    if o == 3:  t=n;n=m;m=t

for i in range(n):
    for j in range(m):
        print(seq[0][i][j],end=' ')
    for j in range(m):
        print(seq[1][i][j],end=' ')
    print()
for i in range(n):
    for j in range(m):
        print(seq[2][i][j],end=' ')
    for j in range(m):
        print(seq[3][i][j],end=' ')
    print()
```
