```python
import sys

num = int(sys.stdin.readline())
#num = 8

pos = [0] * num

flag_row = [False] * num
flag_dia1 = [False] * (num*2 - 1)   # i - j + num- 1   
flag_dia2 = [False] * (num*2 - 1)   # i + j

cnt = 0
def queen(i : int, num : int) :
	global cnt
	for j in range(num) :
		if flag_row[j] == False and flag_dia1[i - j + num - 1] == False and flag_dia2[i + j] == False :
			pos[i] = j
			flag_row[j] = True
			flag_dia1[i - j + num - 1] = True
			flag_dia2[i + j] = True
			if i != num - 1 :
				queen(i + 1, num)
			else : 
				# print(pos[0:])
				cnt += 1
			flag_row[j] = False
			flag_dia1[i - j + num - 1] = False
			flag_dia2[i + j] = False

rst = 0

if num == 1 : rst = 1

elif num % 2 == 0 :
	for j in range(num // 2) :
		flag_row[j] = True
		flag_dia1[0 - j + num - 1] = True
		flag_dia2[0 + j] = True
		queen(1, num)
		flag_row[j] = False
		flag_dia1[0 - j + num - 1] = False
		flag_dia2[0 + j] = False
	rst = cnt * 2

elif num % 2 == 1 :
	for j in range(num // 2) :
		flag_row[j] = True
		flag_dia1[0 - j + num - 1] = True
		flag_dia2[0 + j] = True
		queen(1, num)
		flag_row[j] = False
		flag_dia1[0 - j + num - 1] = False
		flag_dia2[0 + j] = False
	rst = cnt * 2
	cnt = 0
	j = num // 2
	flag_row[j] = True
	flag_dia1[0 - j + num - 1] = True
	flag_dia2[0 + j] = True
	queen(1, num)
	flag_row[j] = False
	flag_dia1[0 - j + num - 1] = False
	flag_dia2[0 + j] = False
	rst += cnt


print(rst)
```

```python
def chk_place(x,y,z,i,q):
    if i in x or i+q in y or i-q in z:
            return False
    return True

def slct_num(x=[],y=[],z=[],q=0):
    global ans
    if q==N:
        ans+=1
    else :
        for i in range(N):
            if chk_place(x,y,z,i,q): #열, 대각선 조건
                x.append(i)
                y.append(i+q)
                z.append(i-q)
                slct_num(x,y,z,q+1)
                x.pop()
                y.pop()
                z.pop()

N=int(input())
ans=0
slct_num([])
print(ans)
```

```python
queen_move=[(0,0),(1,1),(1,0),(1,-1),(0,-1),(-1,0),(-1,1),(0,1)]

def move_queen(i,graph):
    a,b=i%N,i//N
    for k in queen_move:
        for j in range(1,N):
            if 0<=a+j*k[0]<N and 0<=b+j*k[1]<N:
                graph[i+j*k[0]+N*j*k[1]]=1
    return graph

def slct_num(graph,k,q=0): #graph : queen 놓을 수 있는 위치, k = 중복 탐색 제어(for문에서 작동), t= queen놓은 수
    g=[i for i in graph]
    if q == N :
        ans[0]+=1
    elif q < N :
        for i in range(k,k+N):
            if g[i]==0:
                slct_num(move_queen(i,g),N*(q+1),q+1)
                g=[i for i in graph]
                
N=int(input())
n=N**2
graph=[0 for _ in range(n)]
ans=[0]
slct_num(graph,0)
print(ans[0])
```

```python
def find_pri(i):
    r=int(i/2)
    while True:
        if r==i: break
        elif prime_num[r]==1 and prime_num[i-r]==1:return i-r,r
        r+=1
        
n = int(input())
even_ = [int(input()) for _ in range(n)]

e_max = max(even_)
prime_num = [1]*(e_max+1)
prime_num[0:2] = [0,0]

# 소수 list 뽑기
for i in range(2,int(e_max**0.5)+1):
    if prime_num[i]==1:
        for j in range(i+i,e_max+1,i):
            prime_num[j]=0

for i in even_:
    x,y=find_pri(i)
    print(x,y)
```
