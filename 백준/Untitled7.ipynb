```python
##함수 미사용

import datetime
start_time = datetime.datetime.now() 

from collections import deque

k=int(input())
M,N=map(int,input().split())
arr=[list(map(int,input().split())) for _ in range(N)]
    
dv=[(0,1),(1,0),(0,-1),(-1,0)]
dk=[(2,1),(1,2),(-2,1),(-1,2),(2,-1),(1,-2),(-2,-1),(-1,-2)] 

v=[[[-1]*(k+1) for _ in range(M)] for _ in range(N)]
v[0][0][k] = 1
q=deque()
q.append((0,0,0,k))

while q:
    n,x,y,k = q.popleft()

    if x==M-1 and y==N-1: 
        print(n);
        break
       
    for i in dv:
        xx,yy =x+i[0],y+i[1] 
        if 0<=xx<M and 0<=yy<N :
            if arr[yy][xx]!=1:
                if v[yy][xx][k] == -1:
                    v[yy][xx][k] = 1 
                    q.append((n+1,xx,yy,k))
    if k>=1: 
        for i in dk:
            xx,yy =x+i[0],y+i[1]
            if 0<=xx<M and 0<=yy<N :
                if arr[yy][xx]!=1:
                    if v[yy][xx][k-1] == -1:
                        v[yy][xx][k-1] = 1
                        q.append((n+1,xx,yy,k-1))
else : print(-1)



end_time = datetime.datetime.now()
elapsed_time = end_time - start_time
micro_elapsed_time = elapsed_time.microseconds

print(micro_elapsed_time)
731244 263483
990493
```
